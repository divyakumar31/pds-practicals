```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# boundry using basemap 
df = pd.read_csv("../Gujarat_Districts.csv")
df.info()
```

```python
districts = df['District']
populations = df['Population']
coordinates = df['Co-Ordinate']
latitudes, longitudes = zip(*[eval(coord) for coord in coordinates])
```

```python
plt.figure(figsize=(15,10))
m = Basemap(projection='mill', llcrnrlat=20, urcrnrlat=25, llcrnrlon=68, urcrnrlon=75, resolution='f')
m.drawcoastlines()
m.drawlsmask(land_color='lightgreen',ocean_color='aqua')
m.drawcountries()
x, y = m(longitudes, latitudes)
m.plot(x, y, marker='o', markersize=6,markerfacecolor='red', linewidth=0)
for i in range(len(districts)):
    plt.text(x[i], y[i], f"{districts[i]}\n{populations[i]}", fontsize=8, ha='right')
plt.show()
```
