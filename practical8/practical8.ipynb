```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Draw Graph using the following nodes and edges using networkx
nodes = ['1', '2', '3', '4', '5']
edges = [('1', '2'), ('1', '5'), ('2', '3'), ('3', '4'), ('4', '5'), ('1', '4'), ('2', '4')]
```

```python
G = nx.Graph()
```

```python
for n in nodes:
    G.add_node(n)

for e in edges:
    G.add_edge(e[0], e[1])
```

```python
nx.draw(G,with_labels=True,node_color="Green",node_size=1500,font_color="White",width=3)

plt.margins(0.2)
plt.show()
```
