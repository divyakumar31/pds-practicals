Practical 3: Consider company master data of Gujarat dataset. Perform the following operations on it.

a) Groupby the number of records based on the company status attribute and display their statistics. <br>
b) Groupby the number of records based on the company status and company subcategory attributes and display their statistics.

```python
# Import libraries
import pandas as pd
```

```python
df = pd.read_csv("../mca_gujarat_21042018.csv")
```

```python
# Fill the missing values
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
# a) Groupby the number of records based on the company status attribute and display their statistics. 

statistics = df.groupby('COMPANY_STATUS')['COMPANY_STATUS'].value_counts()
statistics.describe()
```

```python
state = df.groupby('COMPANY_STATUS')
state.describe()
```

```python
# b) Groupby the number of records based on the company status and company subcategory attributes and display their statistics.

states = df.groupby(['COMPANY_STATUS', 'COMPANY_SUB_CATEGORY'])
states.describe()
```

```python
states = df.groupby(['COMPANY_STATUS', 'COMPANY_SUB_CATEGORY']).value_counts()
states.describe()
```
