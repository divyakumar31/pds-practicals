```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../loan_data_set.csv')
```

```python
# a) Bar Plot
sns.countplot(data=df, x='Gender', palette='Set1')
plt.title('No. of male and female applicants')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# b) Strip Plot
sns.stripplot(data=df, x='Property_Area', y='ApplicantIncome', palette='Set1')
plt.title('Applicant Income by Property Area')
plt.xlabel('Property Area')
plt.ylabel('Applicant Income')
plt.show()
```
