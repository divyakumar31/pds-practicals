Practical 4: : Consider company master data of Gujarat dataset. Display the statistic regarding the number of records according to the following criteria on the various attributes: 

a) Display the statistics of the company registered in the years 2010 to 2015.<br>
b) Display the statistics of the company whose principal business is finance and registered after 2016.<br>
c) Display the statistics of the company whose principal business is construction and authorized capital is above 10000000.<br>
d) Display the statistics of the company according to the principal business activity and registered in the years 2015 to 2017.

```python
# Import libraries
import pandas as pd
```

```python
df = pd.read_csv('mca_gujarat_21042018.csv')
```

```python
# Fill the missing values
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
# a) Display the statistics of the company registered in the years 2010 to 2015.

# Convert DATE_OF_REGISTRATION column into date format
df['DATE_OF_REGISTRATION'] = pd.to_datetime(df['DATE_OF_REGISTRATION'], format='%d-%m-%Y')
```

```python
# Filter the data for the years 2010 to 2015
company = df[(df['DATE_OF_REGISTRATION'].dt.year >= 2010) & (df['DATE_OF_REGISTRATION'].dt.year <= 2015)]
company.describe()
```

```python
# b) Display the statistics of the company whose principal business is finance and registered after 2016.

company = df[(df['PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN'] == 'Finance') & (df['DATE_OF_REGISTRATION'].dt.year > 2016)]
company.describe()
```

```python
df['PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN'].unique()
```

```python
# c) Display the statistics of the company whose principal business is construction and authorized capital is above 10000000.

company = df[(df['PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN'] == 'Construction') & (df['AUTHORIZED_CAP'] > 10000000)]
company.describe()
```

```python
# d) Display the statistics of the company according to the principal business activity and registered in the years 2015 to 2017.

company = df[(df['DATE_OF_REGISTRATION'].dt.year >= 2015) & (df['DATE_OF_REGISTRATION'].dt.year <= 2017)].groupby('PRINCIPAL_BUSINESS_ACTIVITY_AS_PER_CIN')

company.describe()
```
