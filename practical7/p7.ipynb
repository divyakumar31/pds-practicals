```python
# Import libraries
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd
```

```python
# Let's assume we have the following documents
documents = [
    'This is PDS',
    'PDS-python for data science',
    'This code is in python',
    'Have you learned python?'
]
```

```python
# Bag of Words Model
vectorizer = CountVectorizer()
bow_data = vectorizer.fit_transform(documents)
```

```python
bow_df = pd.DataFrame(data=bow_data.toarray(),columns=vectorizer.get_feature_names_out())
bow_df
```

```python
# TF-IDF Model
vectorizer = TfidfVectorizer()
tf_data = vectorizer.fit_transform(documents)
```

```python
tf_df = pd.DataFrame(data=tf_data.toarray(),columns=vectorizer.get_feature_names_out())
tf_df
```
